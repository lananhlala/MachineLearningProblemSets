```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('ted.csv')

data.head(5)
```

```python
# sort columns
data = data[['name', 'title', 'description', 'main_speaker', 'speaker_occupation', 'num_speaker', 'duration', 'event', 'film_date', 'published_date', 'comments', 'tags', 'languages', 'ratings', 'related_talks', 'url', 'views']]
data.head(5)
```

```python
# convert timestamp to readable
data['film_date'] = data['film_date'].apply(lambda x: datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
data['published_date'] = data['published_date'].apply(lambda x: datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
```

```python
data.head()
```

```python
# Most views TED talk
pop_talks = data[['title', 'main_speaker', 'views', 'film_date']].sort_values('views', ascending=False)[:15]
pop_talks
```

```python
data['views'].describe()
```

```python
# visualize 
plt.figure(figsize=(20,12))
sns.barplot(x='main_speaker', y='views', data=pop_talks)
```

```python
# see distribute views
sns.distplot(data[data['views'] < 4e6]['views'])
```

```python
data['comments'].describe()
```

```python
sns.distplot(data[data['comments'] < 500]['comments'])
```

```python
sns.jointplot(x='views', y='comments', data=data)
```

### Plot the number of talks that took place each year

## Clean the rating data

```python
import ast
```

```python
data['ratings_list'] = data.ratings.apply(lambda x: ast.literal_eval(x))
```

```python
# modify the function to get the sum of count
def get_num_ratings(list_of_dicts):
    num = 0
    for d in list_of_dicts:
        num = num + d['count']
    return num
```

```python
data['num_ratings'] = data.ratings_list.apply(get_num_ratings)
```
